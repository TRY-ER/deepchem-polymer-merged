```python
from validator.validator import StringValidator
```

```python
import pandas as pd


# test_df = pd.read_csv("../datasets/output_obj/lstm_100_gen.csv")
test_df = pd.read_csv("../datasets/output_obj/vae_100_gen.csv")
```

```python
master_df = pd.read_csv("../datasets/finals/dfs/filtered_sequences_case_study.csv")
```

```python
validator = StringValidator()
def validate_sequence(seq):
    try:
        return validator.validate(seq.replace(" ", ""))
    except:
        return False
```

```python
def remove_spaces(seq):
    return seq.replace(" ", "")
```

```python
test_df["decoded_text"] = test_df["decoded_text"].apply(remove_spaces) 
```

```python
test_df["valid"] = test_df["decoded_text"].apply(validate_sequence)
```

```python
# check if sequence is in master df
test_df["in_master"] = test_df["decoded_text"].isin(master_df["sequence"].tolist())
```
