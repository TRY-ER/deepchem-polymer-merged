```python
def get_smiles_split(sequence):
    sequence = sequence.replace(" ", "")
    smiles_part = sequence.split("|")[0]
    # Remove patterns like [5*] using regex
    import re
    smiles_part = re.sub(r'\[\d+\*\]', '', smiles_part)
    parts = smiles_part.split('.')
    assert len(parts) == 2, "The SMILES does not split into two parts as expected."
    return parts
```
