```python
df = pd.read_csv("./dataset-poly_chemprop-poly_type_x5.csv")
```

```python
str_list = df["poly_chemprop_input"].values
```

```python
str_l = map(lambda x : x.split("|")[0], str_list)
```

```python
# write a function using re to remove the wildcard characters like [*:1] where there can be any number between 1-4 as integer
import re


def remove_wildcards(s):
    return re.sub(r'\[\*:\d{1,4}\]', '', s)

str_l = map(remove_wildcards, str_l)
```

```python
str_l = list(str_l)
print(str_l[:10])
```

```python
master_str = "".join(str_l)
```

```python
len(master_str)
```

```python
import re

# write a function to separate sub strings where it can be a character (capital or small), there can be a digit after the character, like A1, a2, B3, b4, etc. which will denote as a single token, remove the parantheses like ("(", ")", "{", "}") and bond sings such as "=" or "#"
def separate_tokens(s):
    # First remove unwanted characters (parentheses, brackets, bond signs, hyphens, plus signs)
    s = re.sub(r'[(){}=#+\[\]\-.]', '', s)
    # Then add spaces around tokens (character optionally followed by digit)
    s = re.sub(r'([A-Za-z]\d?)', r' \1 ', s)
    # Remove standalone digits (not attached to letters)
    s = re.sub(r'\b\d+\b', '', s)
    # Clean up multiple spaces and strip
    s = re.sub(r'\s+', ' ', s).strip()
    return s
```

```python
sep_text = separate_tokens(master_str)
```

```python
u_list = set()
for s in sep_text.split():
    u_list.add(s)
```
