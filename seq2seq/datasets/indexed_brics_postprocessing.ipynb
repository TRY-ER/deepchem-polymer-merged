```python
from rdkit import Chem
```

```python
import joblib
```

```python
indexed_repo = joblib.load("indexed_brics_iter_smiles.pkl")
```

```python
len(indexed_repo)
```

```python
indexed_repo = [item for item in indexed_repo if isinstance(item, str)]
```

```python
filter_repo = set()
```

```python
import re
def get_enclosed(str):
    "this function should get us the part enclosed with ] and ["
    pattern = r'\](.*?)\['
    match = re.search(pattern, str)
    if match:
        return match.group(1)
    return "" 
```

```python
for r in indexed_repo:
    seed, fragement = r.split("|")[2], r.split("|")[4]
    if any(len(get_enclosed(x)) == 1 for x in [seed, fragement]):
        # print("Found just single atom in ", seed, fragement)
        continue
    if seed.count("*") !=2 or fragement.count("*") != 2:
        # print("Found invalid number of candidates", seed, fragement)
        continue
    del seed
    del fragement
    filter_repo.add(r)
```

```python
len(filter_repo)
```

```python
filter_repo = list(filter_repo)
```
