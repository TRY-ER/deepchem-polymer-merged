```python
import pandas as pd
```

```python
df = pd.read_csv("./finals/dfs/loss_involved_df.csv")
```

```python
df.head()
```

```python
df["sequence"].values[:5]
```

```python
block_outro = "<1-2:0.375:0.375<1-1:0.375:0.375<2-2:0.375:0.375<3-4:0.375:0.375<3-3:0.375:0.375<4-4:0.125:0.125<1-3:0.125:0.125<1-4:0.125:0.125<2-3:0.125:0.125<2-4:0.125:0.125"
```

```python
def replace_outro(value):
    splits = value.split("|")
    replace = "|".join(splits[:-1])
    replace += "|"
    return replace
```

```python
replace_outro('[1*]C(=O)C[4*].[4*]CC(O)COC(=O)CCCCCN[5*]|[4*]-[*:1].[5*]-[*:2]>>[$([C&!D1&!$(C=*)]-&!@[#6]):1]-&!@[$([N&!D1&!$(N=*)&!$(N-[!#6&!#16&!#0&!#1])&!$([N&R]@[C&R]=O)]):2] |0.5|0.5|<1-3:0.5:0.5<1-4:0.5:0.5<2-3:0.5:0.5<2-4:0.5:0.5')
```

```python
df["sequence_mod"] = df["sequence"].apply(replace_outro)
```

```python
df = df.drop_duplicates(subset=['sequence_mod'], keep='first')
```

```python
df.shape
```

```python
def add_outro(value):
    return value + block_outro
```

```python
df["sequence_mod"] = df["sequence_mod"].apply(add_outro)
```

```python
df["sequence_mod"].values[69]
```

```python
def cleanup(value):
    value = value.replace('"','').replace(' ', "")
    return value
```

```python
df["sequence_mod"] = df["sequence_mod"].apply(cleanup)
```

```python
df["sequence_mod"].values[69]
```

```python
df.shape
```

```python
df.head()
```

```python
for index,row in df.iterrows():
    if row.is_a_pair == 0:
        penalty = row.logP_delta / 2
    elif row.is_a_pair == 1:
        penalty = row.logP_delta * 2
    df.loc[index,"penalty"] = penalty
```

```python
df.head()
```

```python
col_min = df["penalty"].min()
col_max = df["penalty"].max()
```

```python
df["penalty_norm"] = (df["penalty"] - col_min) / (col_max - col_min)
```

```python
df["penalty_norm"].values
```

```python
df["penalty_norm"] = 1 - df["penalty_norm"]
```

```python
df = df.drop(["sequence", "penalty"], axis=1)
```

```python
df.head()
```

```python
df.to_csv("penalty_max_data.csv", index=False)
```
