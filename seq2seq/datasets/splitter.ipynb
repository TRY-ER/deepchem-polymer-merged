```python
import pandas as pd
```

```python
df = pd.read_csv("dataset-poly_chemprop-poly_type_x5.csv")
```

```python
def generate_split(df):
    """
    This function first creates a column named "mol_id" to assign an id to smiles. then keeps another columns as "smiles" to keep the text form.
    keep an index column with no column name. the columns should be like : ,mol_id,smiles
    other columns should be dropped.

    you have to write three dataframs by splitting the data into train, valid and test in 80%, 10%, 10% respectively.
    the smiles data is saved in column "poly_chemprop_input" in current df
    """
    df = df.rename(columns={"poly_chemprop_input": "smiles"})
    df["mol_id"] = range(len(df))
    df = df[["mol_id", "smiles"]]
    
    # Split the DataFrame into train, valid, and test sets
    train_df = df.sample(frac=0.8, random_state=42)
    remaining_df = df.drop(train_df.index)
    valid_df = remaining_df.sample(frac=0.5, random_state=42)
    test_df = remaining_df.drop(valid_df.index)
    
    # Save the DataFrames to CSV files
    train_df.to_csv("train.csv", index=True)
    valid_df.to_csv("valid.csv", index=True)
    test_df.to_csv("test.csv", index=True)
```

```python
generate_split(df)
```
